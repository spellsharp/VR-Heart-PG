```python
import matplotlib.pyplot as plt
import numpy as np
from monai.transforms import *
```

```python
image_path = "/data/vr_heart/datasets/ImageCHD_dicom/dicoms/ct_1001_image"
label_path = '/data/vr_heart/datasets/ImageCHD_dicom/labels/ct_1001_label.nii.gz'
image = LoadImage(image_only=True)(image_path)
label = LoadImage(image_only=True)(label_path)
```

```python
image_np = image.cpu().numpy()
label_np = label.cpu().numpy()

image_np.shape
```

```python
slice = 199
fig, ax = plt.subplots(1, 2, figsize=(15, 15))
# saggital
ax[0].imshow(image_np[:, :, slice], cmap='gray')
ax[1].imshow(label_np[:, :, slice])
```

```python
slice = 100
fig, ax = plt.subplots(1, 2, figsize=(8, 8))

# coronal
ax[0].imshow(image_np[:, slice, :], cmap='gray')
ax[1].imshow(label_np[:, slice, :])
```

```python
slice = 100
fig, ax = plt.subplots(1, 2, figsize=(8, 8))
# axial
ax[0].imshow(image_np[slice, :, :], cmap='gray')
ax[1].imshow(label_np[slice, :, :])
```
